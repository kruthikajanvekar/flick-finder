```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        counter=0
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
def fetch_director(obj):
    L=[]
    try:
        if pd.isna(obj) or obj == '[]':  # handle null or empty list strings
            return []
        for i in ast.literal_eval(obj):
            if i['job']=='Director':
                L.append(i['name'])
            break
    except:
        return []
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```
